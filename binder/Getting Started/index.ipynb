## 2. Log into your Labstep account using:
### (i) either your apikey

- Authenticating via an API key is the most secure method.

- Video tutorial on how to create an apikey:
https://help.labstep.com/en/articles/3636355-how-to-create-an-api-key-on-labstep

```python
user = labstep.authenticate(username='apitest@labstep.com',
                            apikey='1c5b51d1-a1a5-4661-9eea-f71b7523f909')

# To see attributes of 'user'
print(user)
```

### (ii) or your password

- NOTE: this method will not work for users who sign in via Google or via an Institution.

```python
user = labstep.login(username='apitest@labstep.com',
                     password='apitestpass')

# To see attributes of 'user'
print(user)
```
